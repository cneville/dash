```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
df_spc = yf.download (tickers = "^GSPC", start = "2016-01-01", 
                              end = "2019-09-01", interval = "1d")
```

```python
df_spc.head()
```

```python
# We can use pandas to calculate the moving average for different time periods
df_spc['50ma'] = df_spc['Close'].rolling(window=50).mean()
df_spc['200ma'] = df_spc['Close'].rolling(window=200).mean()
```

```python
df_spc.reset_index(inplace=True)
df_spc = df_spc[['Date', 'Close', '50ma', '200ma']]
df_spc.head()
```

```python
# Let's plot the data to see the moving averages
plt.figure(figsize=(15, 5))
plt.title('S&P 500 Close Price with 200 and 50-Day Moving Average')
plt.plot(df_spc['Close'])
plt.plot(df_spc['50ma'])
plt.plot(df_spc['200ma'])
plt.legend(['Close', '50MA', '200MA'])
plt.show()
```
